```python
import numpy as np
import pandas as pd
heroes = pd.read_csv("data/heroes_information.csv")
```

```python
###instalar desde anaconda navigator######
from pandasql import *
```

```python
pysqldf = lambda q: sqldf(q,globals())
```

```python
query2 = " SELECT * FROM heroes"
```

```python
pysqldf(query2)
```

# 1) ¿Cuál es el nombre del primer superhéroe de la tabla heroes?

```python
query_1 = " SELECT * FROM heroes"
pysqldf(query_1)
```

**Respuesta:** A-Bomb

# 2) ¿Cuántas casas publicadoras existen?

```python
query_2 = "SELECT COUNT(DISTINCT Publisher) FROM heroes"
pysqldf(query_2)
```

**Respuesta:** 24

# 3) ¿Cuántos superhéroes miden más de 2 metros (200 cms)

```python
query_3 = "SELECT COUNT(*) FROM heroes WHERE Height > 200"
pysqldf(query_3)
```

**Respuesta:** 59

# 4) ¿Cuántos superhéroes son humanos y miden más de 2 metros (200 cms)?

```python
query_4 = "SELECT COUNT(*) FROM heroes WHERE Height > 200 AND Race = 'Human'"
pysqldf(query_4)
```

**Respuesta:** 12

# 5 ¿Cuántos superhéroes pesan más de 100 lbs y menos de 200 lbs?

```python
query_5 = "SELECT Count(*) FROM heroes WHERE Weight > 100 AND Weight < 200 "
pysqldf(query_5)
```

**Respuesta:** 98

# 6) ¿Cuántos superhéroes tienen los ojos de color azul o rojo?

```python
query_6 = "SELECT Count(*) FROM heroes WHERE `Eye color` = 'red' OR `Eye color` = 'blue'"
pysqldf(query_6)
```

**Respuesta:** 271

# 7) ¿Cuántos superhéroes son Human, Mutant y tienen el pelo color Verde o son Vampiros con pelo Negro?

```python
query_7 = " SELECT COUNT(*) FROM heroes WHERE  (Race IN ('Human','Mutant') AND `Hair color` = 'Green') OR  (Race = 'Vampire' AND `Hair color` = 'Black')"
pysqldf(query_7)
```

**Respuesta:** 5

# 8) ¿Cuál es el nombre del primer superhéroe si ordeno la tabla por raza en orden descendente?

```python
query_8 = " SELECT * FROM heroes ORDER BY Race DESC"
pysqldf(query_8)
```

**Respuesta:** Solomon Grundy

# 9) ¿Cuántos superhéroes son de género masculino y cuántas de género femenino?

```python
query_9 = "SELECT Gender,COUNT(*) FROM heroes GROUP BY Gender" # `Hair color`
pysqldf(query_9)
```

**Respuesta:** Female = 200 &nbsp; | &nbsp; Male = 505

# 10) ¿Cuántas casas publicadoras tienen más de 15 superhéroes?

```python
query_10 = " SELECT COUNT(*) FROM (SELECT COUNT(*) FROM heroes GROUP BY Publisher HAVING COUNT(*) > 15)"
pysqldf(query_10)
```

**Respuesta:** 4
